## Part II. GDP data from EUROSTAT

```python
# import the data files you downloaded from EUROSTAT

df_gdp = pd.read_csv (r'C:/Users/danie/Documents/FEP/Doutoramento/Econometria_II/homeworks/econ108-hw1/data/gdp_data.csv', usecols=[7,8])
```

```python
# create a dataframe with a datetime index matching the timing of the series you downloaded

df_gdp['date'] = pd.PeriodIndex(df_gdp['TIME_PERIOD'], freq='Q').to_timestamp()
df_gdp.set_index('date',inplace=True)
df_gdp.drop(['TIME_PERIOD'], axis=1, inplace=True)
df_gdp.rename(columns={"OBS_VALUE": "GDP"}, inplace=True)
```

```python
# save as a csv file under the name "Real-GDP-EA.csv"

df_gdp.to_csv(f'data/Real-GDP-EA.csv', index=True)
```
